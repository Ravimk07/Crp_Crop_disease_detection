```python
import matplotlib.patches as mpatches

from skimage.filters import threshold_otsu
from skimage.segmentation import clear_border
from skimage.measure import label, regionprops
from skimage.morphology import closing, square
from skimage.color import label2rgb

# apply threshold
thresh = threshold_otsu(image_slice_V)
bw = closing(image_slice_V > thresh, square(3))

# remove artifacts connected to image border
cleared = clear_border(bw)

# label image regions
label_image = label(cleared)
image_label_overlay = label2rgb(label_image, image=image_slice_V)


f = plt.figure(figsize=(10,3))
ax = f.add_subplot(121)
ax2 = f.add_subplot(122)
ax.imshow(image1)
ax2.imshow(cleared)
```

```python
def mask_color_img(img, mask, color=[255,  255], alpha=0.3):
 
    out = img.copy()
    img_layer = img.copy()
    img_layer[mask] = color
    out = cv2.addWeighted(img_layer, alpha, out, 1 - alpha, 0, out)
    return(out)

ff = mask_color_img(image1,cleared,color=[0, 0, 255], alpha=0.5)

f = plt.figure(figsize=(14,3))
ax = f.add_subplot(121)
ax.imshow(image1)
ax = f.add_subplot(122)
ax.imshow(ff)
```

```python
from skimage import filters, segmentation

# find a dividing line between 0 and 255
# pixels below this value will be black
# pixels above this value will be white
val = filters.threshold_otsu(image_slice_V)

# the mask object converts each pixel in the image to True or False
# to indicate whether the given pixel is black/white
mask = image_slice_V > val

# apply the mask to the image object
clean_border = segmentation.clear_border(mask)

# plot the resulting binarized image
plt.imshow(clean_border, cmap='gray')
plt.show()
```
